```python
import numpy as np

a = np.array([1.7, 4.0, 3.1])

exp_a = np.exp(a)    #지수 함수 (분모)
print(exp_a)

sum_exp_a = np.sum(exp_a)  #지수 합수의 합 (분자)
print(sum_exp_a)

y = exp_a / sum_exp_a    #소프트맥스 함수 식
print(y)
```

```python
def softmax(x):
    exp_x = np.exp(x)
    sum_exp_x = np.sum(exp_x)
    y = exp_x / sum_exp_x
    
    return y
```
